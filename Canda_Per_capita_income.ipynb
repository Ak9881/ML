```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
df
```

```python
df = df.rename(columns={'per capita income (US$)': 'per_capita'})
```

```python
df
```

```python
plt.xlabel("year")
plt.ylabel("per_capita")
plt.scatter(df.year,df.per_capita, color ='Red',marker='+')

```

```python
df_year = df.drop("per_capita", axis='columns')
```

```python
df_year
```

```python
df
```

```python
per_capita = df.per_capita
per_capita
```

```python
from sklearn import linear_model
```

```python

# Create linear regression object
reg = linear_model.LinearRegression()
reg.fit(df_year,per_capita)
```

```python
reg.predict([[2000]])
```

```python
reg.predict([[2021]])
```

```python
reg.predict([[2022]])
```

```python
plt.xlabel("year")
plt.title("Canada Per capita income (Y o Y)")
plt.ylabel("per_capita")
plt.scatter(df.year, df.per_capita, color = 'Red', marker='+')
plt.plot(df.year,reg.predict(df[['year']]), color = 'red')
```
